```python
df = dataframe.copy()
```

```python
movie_features = df.pivot_table(index = 'program_name', columns='user_id_maped', values='rating').fillna(0)
movie_features.head()
```

```python
"""
TODO build Recommender system to Highlight Programs that usesrs might be interested in
"""
```

```python
from scipy.sparse import csr_matrix
movie_features_df_matrix = csr_matrix(movie_features.values)
```

```python
from sklearn.neighbors import NearestNeighbors
model_knn = NearestNeighbors(metric='cosine', algorithm='brute')
model_knn.fit(movie_features_df_matrix)
```

```python
"""
TODO show the recommendations (top 5) for the people who watched "Moana" movie
"""
```

```python
movie_features_df_selection= movie_features.reset_index()
movie_features_df_selection[['program_name']]
df[['program_name']].head(10)
```
